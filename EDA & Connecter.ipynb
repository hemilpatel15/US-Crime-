```python
from sqlalchemy import create_engine
# Standardize column names
df.columns = [col.strip().lower().replace(" ", "_").replace("-", "_").replace("/", "_") for col in df.columns]

# Rename specific column that caused error
df.rename(columns={'street_state': 'street_state'}, inplace=True)

# Format date columns
df['date_rptd'] = pd.to_datetime(df['date_rptd'], errors='coerce').dt.date
df['date_occ'] = pd.to_datetime(df['date_occ'], errors='coerce').dt.date

# Connect to MySQL
engine = create_engine("mysql+mysqlconnector://root:12345678@localhost:3306/crime_db")

# Upload to database
df.to_sql('crime_reports', con=engine, if_exists='append', index=False)

print("✅ Upload successful!")
```

```python
from sqlalchemy import text

# Example 1: Get total number of crimes
with engine.connect() as conn:
    result = conn.execute(text("SELECT COUNT(*) FROM crime_reports"))
    print("Total records:", result.scalar())

# Example 2: Top 5 crime types
query = """
SELECT crm_cd_desc, COUNT(*) AS count 
FROM crime_reports
GROUP BY crm_cd_desc
ORDER BY count DESC
LIMIT 5;
"""
pd.read_sql(query, engine)
```
